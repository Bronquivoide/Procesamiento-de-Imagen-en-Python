# Ejercicio 1:

### Explorar las componentes rojo, verde y azul de la famosa imagen usada en otros notebooks, *Verduras.jpg*.

### Primeramente vamos a hacer un peque침o an치lisis de la imagen:


```python
import numpy as np
import matplotlib.pyplot as plt

im=plt.imread('Verduras.jpg')
plt.imshow(im)     
plt.title('Verduras')
plt.show()
```

```python
print(f"El arreglo asociado a la imagen es:\n {im}")
print("")

print(f"El orden del arreglo es:\n {im.shape}")
print("")

print(f"El tipo de arreglo es:\n {type(im)}")
print("")

print(f"La dimensi칩n del arreglo NumPy es:\n {im.ndim}")
print(" ")
```

### Del comando anterior vemos que la funci칩n *plt.imshow()* recibe tres matrices (una por componente) representadas en un arreglo tridimensional NumPy asociado a las componentes de color de la imagen para mostrarla.

### Podemos observar la imagen separ치ndola por componentes. Vamos a extraer cada una de las componentes rojo, verde y azul de la imagen, y a mostrarlas individualmente. Para esto vamos a usar la funci칩n *plt.imshow()* aplicada sobre cada componente extra칤da.

### Para extraer cada componente usamos la funci칩n *imagen[ largo , ancho , componente (o canal) ]*. Como queremos 칰nicamente mostrar cada componente a trav칠s de todo el largo y el ancho, vamos a establecer la funci칩n como *imagen[ : , : , $i$ ]*, en donde $i=0,1,2$ (rojo, verde, azul).

```python
#R
plt.imshow(im[:,:,0])    
plt.title('Componente roja de Verduras.jpg',fontsize=20,color="r")
plt.show()

#G
plt.imshow(im[:,:,1])    
plt.title('Componente verde de Verduras.jpg',fontsize=20,color="g")
plt.show()

#B
plt.imshow(im[:,:,2])    
plt.title('Componente azul de Verduras.jpg',fontsize=20,color="b")
plt.show()
```

### O bien, podemos visualizar las componentes en escala de grises, donde los valores de intensidad asociado a los pixeles de los canales RGB se interpretan como niveles de gris. Es decir,  los valores num칠ricos de la intensidad en RGB se interpretan en el rango de niveles de gris.

```python
#R
plt.imshow(im[:,:,0],cmap='gray')    
plt.title('Componente roja de Verduras.jpg en escala de grises',fontsize=15,color="r")
plt.show()

#G
plt.imshow(im[:,:,1],cmap='gray')    
plt.title('Componente verde de Verduras.jpg en escala de grises',fontsize=15,color="g")
plt.show()

#B
plt.imshow(im[:,:,2],cmap='gray')    
plt.title('Componente azul de Verduras.jpg en escala de grises',fontsize=15,color="b")
plt.show()
```
