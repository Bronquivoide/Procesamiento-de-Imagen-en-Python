### Repasar los fundamentos básicos de las imágenes RGB con la librería Open CV. Para esto vamos a trabajar con la imagen *Leopardo.jpg* implementando los procesos de lectura de imagen y de mostrarla.

### Primeramente observemos la principal diferencia entre leer la imagen con Matplotlib y luego con OpenCV (en ambos casos la imagen se va a mostrar con Matplotlib):

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_matplotlib=plt.imread('Leopardo.jpg')
plt.imshow(img_matplotlib)     
plt.title('Leopardo.jpg en Matplotlib', fontsize=20)
plt.show()

img_cv = cv2.imread('Leopardo.jpg')
plt.imshow(img_cv)
plt.title('Leopardo.jpg en OpenCV', fontsize=20)
plt.show()
```

### Notamos que Matplotlib, como lo hemos mencionado en notebooks anteriores, muestra la imagen de color en el espacio RGB, mientras que OpenCV la muestra en el espacio BGR.

### Una forma de esta diferencia es con base a la esquematización del espacio tridimensional RGB presentada en el notebook de fundamentos de imágenes RGB y su analogía con el espacio $\mathbb{R}^3$. Podríamos decir que con respecto a la base que genera a RGB, OpenCV toma por default una base que genera al espacio BGR para representar a la imagen. Digamos que BGR toma las componentes de distinta manera bajo la regla: $\{R_{RGB} \rightarrow B $ , $B_{RGB} \rightarrow R \}$.

### Para mostrar la imagen con OpenCV en términos del espacio RGB usamos el siguiente comando:

```python
imagenRGB= cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)  #Esta función convierte el formato de una imagen del espacio BGR al espacio RGB.
plt.title('Leopardo.jpg en OpenCV en formato RGB', fontsize=20)
plt.imshow(imagenRGB)
plt.show()
```

### O bien, podemos mostrar la misma imagen, pero con la magnitud de las intensidades (asociadas a cada pixel) equivalente a la magnitud de las intensidades en escala de grises:

```python
imagen_grises= cv2.cvtColor(imagenRGB, cv2.COLOR_BGR2GRAY)
plt.imshow(imagen_grises, cmap = 'gray')
plt.title('Leopardo.jpg en OpenCV en escala de grises', fontsize=20)
plt.show()
```

### O para fines prácticos podemos leerla y mostrarla en escala de grises de una:

```python
imagen_grises_de_una= cv2.imread('Leopardo.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_grises_de_una, cmap = 'gray')
plt.show()
```
