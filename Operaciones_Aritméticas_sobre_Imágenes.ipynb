# Introducción:

### Recordemos que la intensidad de un pixel en las coordenadas espaciales $(x_0,y_0)$ en una imagen arbitraria $f$ (cuyo arreglo de datos de intensidades que la representa es del orden de $M \times N$), está dada por la magnitud:

<font size="5"> $f(x_0,y_0)=I$, $\forall x_0 \in \mathbb{N}_{[0,M]}$ y $\forall y_0 \in \mathbb{N}_{[0,N]}$. </font>

#### <span style="font-family: 'Verdana'; color: blue;"> Nota: En este notebook definimos el conjunto de los naturales de $0$ a $M$ como $\mathbb{N}_{[0,M]}=\{0,1,...,M\}$ y el conjunto de los naturales de $0$ a $N$ como $\mathbb{N}_{[0,N]}=\{0,1,...,N\}$, esto para denotar que las coordenadas espaciales de los pixeles son valores enteros positivos o cero, pues toman magnitudes desde cero hasta el orden del tamaño de la imagen. (Aquello de incluir al cero en el conjunto bajo esta notación es para unificar sentimientos sobre aquellos que no consideran al cero natural con los que sí).  </span> 


### También recordemos que las intensidades de los pixeles se encuentran en el rango $0 \leq I \leq 255$, o equivalentemente, $0 \leq I \leq 1$, en donde $[0,255]\subset \mathbb{Z}^+ \cup \{0\} := \mathbb{Z} _{\geq 0}$, y, $ [0,1]\subset \mathbb{R}^+ \cup \{0\}:= \mathbb{R} _{\geq 0}$.

### De lo anterior podemos definir el mapeo de $f$ mediante:

<font size="5"> $f: \mathbb{N}_{[0,M]} \times \mathbb{N}_{[0,N]} \rightarrow [0,255] \subset \mathbb{Z} _{\geq 0}$ </font>  

### O bien:

<font size="5"> $f: \mathbb{N}_{[0,M]} \times \mathbb{N}_{[0,N]} \rightarrow [0,1] \subset \mathbb{R} _{\geq 0}$ </font>

# Operación aritmética:

### Definimos una imagen resultante de una operación aritmética efectuada entre dos imágenes arbitrarias bajo la relación:

<font size="5"> $R=I_1 \odot I_2$ </font>

### La operación $\odot$ está definida entre los arreglos asociados a las imágenes del mismo orden $M \times N$, por lo que la imagen resultante $R$ está representada por un arreglo de este orden.

# Definiendo las operaciones aritméticas:

### Sean $f(x,y)$ y $g(x,y)$ imágenes arbitrarias, definimos las siguientes cuatro operaciones que proporcionan una imagen resultante:

### $\blacksquare$ Suma: $R_s(x,y)=f(x,y)+g(x,y)$

### $\blacksquare$ Resta: $R_r(x,y)=f(x,y)-g(x,y)$

### $\blacksquare$ Multuplicación: $R_m(x,y)=f(x,y) \cdot g(x,y)$

### $\blacksquare$ División: $R_d(x,y)=f(x,y)/g(x,y)$

### Las operaciones están definidas entrada por entrada entre los arreglos. (Nótese como mañosamente hemos abordado al conjunto de datos que representa a la imagen con los conceptos *arreglo* y *matriz*, en este notebook, por ejemplo, no hemos abordado este último, como cuando se describieron las imágenes RGB bajo el caracter del espacio RGB como espacio vectorial).

### Hablando por ejemplo del producto, probablemente va a hacer ruido que el producto sea entre una matriz $M \times N$ y otra del mismo orden, pues para empezar el producto matricial está definido entre matrices de $M \times N$ y de $N \times P$, además, no está definido entrada por entrada. Mismo caso para la división, empezando con que no tiene sentido hablar de divisiones entre matrices, sino de la multiplicación de una matriz por una inversa.

### Y es que cuando tratamos operaciones aritméticas entre imágenes, hablamos de operaciones entre arreglos definidas entrada por entrada, no de operaciones matriciales. Las operaciones entre arreglos son entre intensidades de pixeles en pares coordenados $(x,y), \forall x=0,1,2,...,M-1,M$ y $y=0,1,2,...,N-1,N$.

# Ejercicio:

### Implementar con OpenCV las cuatro operaciones aritméticas sobre la imagen *Lena.tif*, la cual está en escala de grises.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

im=cv2.imread("Lena.tif")
print(im.shape)

#Habíamos mencionado que se carga la imagen con OpenCV en tres componentes de color a pesar de estar en escala de grises, por lo que 
#vamos a cargar la imagen original (en escala de grises) mediante el siguiente comando:

imagen= cv2.imread('Lena.tif', 0)
print(imagen.shape)
plt.figure(figsize=(12,6))
plt.imshow(imagen, cmap = 'gray')
plt.title("Lena.tif",fontsize=25)
plt.show()
print(f"El arreglo de datos asociado a Lena.tif es:\n {imagen}")
print("")
print(f"Del tipo:\n{type(imagen)}")
print("")
print(f"Del orden:\n{imagen.shape}")
print("")
print(f"Cuyo formato de datos es:\n{imagen.dtype}")
```

# $\blacksquare$ Adición sobre Lena.tif:

### Lo que hay que hacer un arreglo del mismo orden que Lena.tif para operar la suma entre estos.

### Consideremos el arreglo de $256 \times 256$ en donde cada entrada es el escalar $100$, que digamos podría representar a una segunda imagen, y procederemos a operar la suma entre el arreglo asociado a Lena.tif y al segundo arreglo mencionado.


```python
#Generando el arreglo:
arreglo= np.ones(imagen.shape, dtype = "uint8")*100
print(f"El arreglo es:\n {arreglo}")

#Adición entre la imagen y el arreglo:
suma= cv2.add(imagen,arreglo)
suma

#Mostrando la imagen:
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(imagen, cmap = 'gray')
plt.title("Imagen original", fontsize = 25)
plt.subplot(1,2,2)
plt.imshow(suma , cmap = 'gray')
plt.title("Imagen bajo el operador suma", fontsize = 25)
plt.show()

print(f"El arreglo asociado a la suma es:\n {suma}")
```

### Nótese como la función *cv2.add()* asigna los valores de las intensidades. Por ejemplo, en la entrada $(x=0,y=0)$ asociada al arreglo de Lena.tif es 162, mientras que en el arreglo es 100, recordando que la escala de intensidades es el intervalo $[0,255]$ de 256 valores posibles, *cv2.add()* fija el máximo posible a $255$. 

### De esta manera el valor 262 se asigna automáticamente a 255:

```python
print(f"En Lena.tif, la entrada (0,0) toma el valor: {imagen[0][0]}")
print()
print(f"En el arreglo, la entrada (0,0) toma el valor:{arreglo[0][0]}")
print()
print(f"La suma de estas dos cantidades es: {162+100}")
print()
print(f"La función cv2.add arroja que la suma es:{suma[0][0]}")
```

### Lo anterior no pasaría con el operador típico de suma de NumPy. NumPy más bien normaliza valores de intensidades $I>255$ al total de valores posibles bajo la diferencia $I-256$, mostrando la imagen con esta diferencia de datos en el arreglo:

```python
suma_np = imagen+arreglo
plt.figure(figsize=(12,6))
plt.imshow(suma_np, cmap = 'gray')
plt.title("Suma bajo el operador de NumPy", fontsize = 15)
plt.show()
print(suma_np)

```

# $\blacksquare$ Substracción sobre Lena.tif:

### Seguimos el procedimiento análogamente al de la adición (centrándonos en OpenCV, pues sucede igual con NumPy):

```python
#Resta entre Lena y el arreglo:
resta= cv2.subtract(imagen, arreglo)

#Comparativa:
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(imagen, cmap = 'gray')
plt.title("Imagen original", fontsize = 25)
plt.subplot(1,2,2)
plt.imshow(resta , cmap = 'gray')
plt.title("Imagen bajo el operador resta", fontsize = 25)
plt.show()

print(f"El arreglo asociado a la resta es:\n {resta}")
```

# $\blacksquare$ Multiplicación sobre Lena.tif:

### Para este caso consideremos la multiplicación por escalar, no como tal por el arreglo usado en la suma y en la resta. Vamos a multiplicar el arreglo de Lena.tif por el escalar 2:

```python
multiplicación= cv2.multiply(imagen, 2)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(imagen, cmap = 'gray')
plt.title("Imagen Original", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(multiplicación , cmap = 'gray')
plt.title("Imagen bajo el operador multiplicación por escalar", fontsize = 15)
plt.show()

print(f"El arreglo asociado a la multiplicación es:\n {multiplicación}")
```

# $\blacksquare$ División sobre Lena.tif:

### Mismo caso que para la multiplicación. Consideremos la división entre el escalar 20 (o la multiplicación por $\frac{1}{20}$):

```python
división= cv2.divide(imagen, 20)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(imagen, cmap = 'gray')
plt.title("Imagen original", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(división , cmap = 'gray')
plt.title("Imagen bajo el operador división por escalar", fontsize = 15)
plt.show()
```
