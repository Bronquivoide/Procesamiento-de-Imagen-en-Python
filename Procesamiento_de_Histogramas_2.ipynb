# $\blacksquare$ Histograma de la componente roja:

```python
#Histograma rojo:
rojo=cv2.calcHist([imagen_RGB], channels = [0], mask = None, histSize = [256], ranges = [0,256])
plt.figure(figsize=(14,7))
plt.plot(rojo, "r")
plt.title("Histograma de la componente roja", fontsize = 25,color="r")
plt.xlabel(r"$r_k$", fontsize = 20,color="r")
plt.ylabel(r"$H(r_k)=n_k$", fontsize = 20,color="r")
plt.show()

#Datos del histograma rojo:
print(f"El orden del histograma de la componente roja es: {rojo.shape}")
print(" ")
print(f"En la componente roja, el n√∫mero m√°ximo de pixeles que comparten el mismo valor de intensidad es {int(np.max(rojo))}, cuya intensidad corresponde a {np.argmax(rojo)}.")
print(" ")
print(f"En la componente roja, el n√∫mero m√≠nimo de pixeles que comparten el mismo valor de intensidad es {int(np.min(rojo))}, cuya intensidad corresponde a {np.argmin(rojo)}.")

#Histograma rojo normalizado:
rojo_normalizado=(rojo) / (imagen_RGB.size) #la funci√≥n .size considera el producto MxN del orden de la imagen
plt.figure(figsize=(14,6))
plt.plot(rojo_normalizado,color="r")
plt.xlabel(r"$r_k$", fontsize = 20,color="r")
plt.ylabel(r"$P(r_k)$", fontsize = 20,color="r")
plt.title("Histograma de la componente roja normalizado", fontsize=25,color="r")
plt.show()

#Datos del histograma rojo normalizado:
print(f"La probabilidad m√°xima es de {np.max(rojo_normalizado)}, la cual corresponde a hallar un nivel de intensidad de {np.argmax(rojo_normalizado)}.")
print(" ")
```
