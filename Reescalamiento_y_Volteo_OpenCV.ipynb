## $\blacksquare$ *Resize*:

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imagenBGR=cv2.imread('Leopardo.jpg')
imagen= cv2.cvtColor(imagenBGR, cv2.COLOR_BGR2RGB)

print(f"La forma de la imagen (su orden, su tamaño, o el tamaño del arreglo que la representa en magnitud de número de pixeles) es de:\n {imagen.shape}")
print("")

plt.imshow(imagen)
plt.title(r'Leopardo.jpg en formato RGB de $480 \times 640$', fontsize=20)
plt.show()
```

### Al arreglo de tres canales de color de $480 \times 640$ asociado a la imagen, vamos a reescalarlo al orden, por ejemplo, de $500 \times 350$ con la función *resize*:

```python
resize_imagen=cv2.resize(imagen,(500,350)) 

print(f"Ahora la forma de la imagen es de:\n {resize_imagen.shape}")
print("")

plt.imshow(resize_imagen)
plt.title(r'Leopardo.jpg de $350 \times 500$', fontsize=20)
plt.show()
```

## $\blacksquare$ *Flipping*:

```python
#Definimos el volteo de la imagen original a lo largo del eje X mediante:
volteoX = cv2.flip(imagen,0)
plt.imshow(volteoX)
plt.show()

#Ahora, definimos el volteo de la imagen original a lo largo del eje Y mediante:
volteoY = cv2.flip(imagen,1)
plt.imshow(volteoY)
plt.show()

#Definimos un volteo simultáneo a lo largo de los dos ejes mediante:
volteoXY = cv2.flip(imagen,-1)
plt.imshow(volteoXY)
plt.show()
```

### Comparando más detalladamente:

```python
plt.figure(figsize=(12,8))

#Original
plt.subplot(2,2,1)
plt.imshow(imagen)
plt.axis('off')
plt.title(" Imagen original", fontsize = 15)

#X
plt.subplot(2,2,2)
plt.imshow(volteoX)
plt.axis('off')
plt.title("Volteo con respecto al eje $X$", fontsize = 15)

#Y
plt.subplot(2,2,3)
plt.imshow(volteoY)
plt.axis('off')
plt.title("Volteo con respecto al eje $Y$", fontsize = 15)

#X,Y
plt.subplot(2,2,4)
plt.imshow(volteoXY)
plt.axis('off')
plt.title("Volteo con respecto a ambos ejes simultáneamente", fontsize = 15)
plt.show()
```
