# Introducción:

### Anteriormente trabajamos con transformaciones espaciales sobre imágenes, las cuales mencionamos que afectan únicamente a la estructura coordenada de los pixeles, no a su intensidad asociada. Trabajamos con transformaciones de la forma:

<font size="5"> $T[f(x,y)]=g(x,y) $</font>

<font size="5"> $\sim $</font>

<font size="5"> $T\begin{pmatrix}
x \\
y  
\end{pmatrix}=\begin{pmatrix}
x' \\
y'  
\end{pmatrix} $</font>
### En donde $f(x,y)$ es una imagen que al aplicarle la transformación espacial $T$, se arroja una imagen de salida $g(x,y).$ 

### En esta serie de notebooks vamos a trabajar ahora con transformaciones de intensidad (también llamadas transformaciones de niveles de grises).

### Consideremos que en imagen $f$ a un pixel coordenado arbitrario $(x_0,y_0)$, se le asocia una intensidad $r$; y bajo las mismas condiciones, en la imagen de salida $g$ se asocia una intensidad $s$. En términos de estas intensidades, las transformaciones ahora son de la forma:

<font size="5"> $T(r)=s $</font>

### Es decir, se aplica $T$ para cada pixel de $f$, generando cada pixel de $g$. $T$ produce un contraste distinto a la imagen original que puede dar información que no se note en ella.

### Sobre las transformaciones de intensidad, hay algunas básicas que trabajaremos en esta seriación de notebooks. En esta primera parte vamos a empezar con la transformación de negativos de imagen.

# Negativos de imagen:

### Supongamos que la intensidad $r$ asociada a los pixeles de una imagen $f$ toma valores acotados por una constante $L$ tal que $0 \leq r \leq L-1$. El negativo de la imagen de intensidad $r$ se obtiene mediante una transformación negativa:

<font size="5"> $T(r)=L-1-r=s $ </font>

### Se dice que $T$ permite un contraste inverso, invirtiendo los niveles de intensidad de la imagen tal que produce un equivalente correspondiente a una imagen negativa.

### Como se puede observar, la relación $r-s$ es lineal decreciente (pues tiene pendiente $-1$) con raíz en $L-1$, pues $T(r_0=L-1)=0=s|_{r_0}$. 

### La constante $L$ es el número de posibles intensidades. Por ejemplo, si la intensidad máxima presente en una imagen es el máximo, $255$, $L$ es la cardinalidad del conjunto $\{0,1,...,254,255\}$, es decir, $L=256$. Si la intensidad máxima presente es de $200$, $L$ es la cardinalidad de $\{0,1,...,200\}$, es decir, $L=201$.

# Ejercicio:

### Supongamos que queremos usar esta transformación sobre una radiografía digital de un cráneo tal que se pueda obtener información estructural del cráneo que puede no verse en el método de imagen original. 

### Vamos a aplicar una transformación de negativos de imagen, usando OpenCV a la imagen *"Cráneo.jpg"*. (Ojo: Cráneo.jpg al final se renombró sin la tilde, Craneo.jpg, pues OpenCV se queja de que los nombres de las imágenes lleven tildes).

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

radiografía=cv2.imread('Craneo.jpg', 0)
plt.figure(figsize=(12,6))
plt.imshow(radiografía, cmap = 'gray')
plt.show()

#Datos sobre los pixeles de la radiografía:
print(f"El tamaño de la imagen es: {radiografía.shape}")
print(" ")
print(f"El valor máximo de intensidad asociado a la imagen es: {np.max(radiografía)}")
print(" ")
print(f"El valor mínimo de intensidad asociado a la imagen es: {np.min(radiografía)}")

#Definiendo e implementando la transformación de negativos:
#Tomemos en cuenta que como r_max = 255, L=256, entonces L-1=255

radiografía_negativa=abs(255 - radiografía)

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title("Radiografía (imagen original)", fontsize = 15)
plt.imshow(radiografía, cmap = 'gray')
plt.subplot(1,2,2)
plt.title("Imagen negativa", fontsize = 15)
plt.imshow(radiografía_negativa, cmap = 'gray')
plt.show()
```

### Notemos las diferencias entre la información estructural de una y otra imagen. En la imagen negativa parece haber estructuras más notorias a lo largo del cráneo que presentan un contraste más oscuro, posiblemente sean nervios que no vemos con tanta facilidad en la radiografía original.
