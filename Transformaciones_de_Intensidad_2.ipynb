# Transformación logaritmo:

### En su forma general, esta transformación está definida mediante:

<font size="5"> $T(r)=c\cdot log(1+r)=s $ </font>

### En donde habíamos mencionado que $r$ es la intensidad asociada a los pixeles de la imagen de entrada, $s$ la intensidad asociada a los pixeles de la imagen de salida y $c$ es una constante positiva que cuando se trabaja con algoritmos computacionales, normalmente está definida por:

<font size="5"> $c= \frac{255}{log(1+r_{max})} $ </font>

### Así mismo, recordemos que $0 \leq x \leq L-1$, en donde $L$ era la cardinalidad del conjunto $\{ r_{min} , \dots , r_{max} \}$.

### Notemos que $s_{min}=T(r=0)=0$ y $s_{max}=T(r=L-1)=c\cdot log(1+(L-1))$. También vemos que conforme $r \rightarrow 0$, entonces $s \rightarrow 0$; y así mismo, conforme  $r \rightarrow L-1$, $s \rightarrow L-1$.

# Ejercicio:

### Implementar una transformada logaritmo a la imagen *"Luna.jpg"* usando OpenCV2.



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#Display de Luna.jpg
imagen=cv2.imread('luna.jpg',0)
plt.figure(figsize=(12,6))
plt.imshow(imagen, cmap = 'gray')
plt.title("Luna.jpg",fontsize=20)
plt.show()

#Definimos la constante c con base a la intensidad máxima de la imagen de enetrada:
print(f"La intensidad máxima asociada a Luna.jpg es: {np.max(imagen)}")
print()
print(f"La intensidad mínima asociada a Luna.jpg es: {np.min(imagen)}") #ojo que el programa va a detectar divisiones entre cero, pues log(1)=0
c= 255 / (np.log(1+np.max(imagen)))
print()
print(f"c={c}")
print()

#Implementamos la tranformación logaritmo:
Tlog=c*(np.log(imagen + 1))
print(f"El arreglo asociado a la imagen de salida es:\n {Tlog}")
print()
print(f"Cuyo tipo de datos es: {Tlog.dtype}") #log entonces float

#Intensidades enteras:
Tlog_sobre_luna=np.array(Tlog, dtype=np.uint8)

plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("Imagen original", fontsize = 15)
plt.imshow(imagen, cmap = 'gray')
plt.subplot(1,2,2)
plt.title("Imagen de salida que arroja la transformación logaritmo ", fontsize = 15)
plt.imshow(Tlog_sobre_luna, cmap = 'gray')
plt.show()

```
