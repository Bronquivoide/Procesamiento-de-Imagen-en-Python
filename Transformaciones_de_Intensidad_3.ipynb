### Tomando como punto de partida las dos primeras partes de esta serie de notebooks, en esta tercera nos corresponde abordar las transformaciones gamma de intensidad.

### En su forma general, las transformaciones gamma se escriben como:

<font size="5"> $T(r)=c\cdot r^{\gamma}=s $ </font>

### En donde $0 \leq r \leq L-1$ es la intensidad asociada a los pixeles de la imagen de entrada y $0 \leq s \leq L-1$ la intensidad asociada a los pixeles de la imagen de salida.

### Las constantes $c$ y $\gamma$ son positivas. A $c$ se le conoce como el factor de escala, el cual normaliza el valor máximo de intensidad, mientras que a $\gamma$ se le llama el factor de corrección.

### Las transformaciones gamma tienen su fundamento en corregir intensidades de imágenes que pueden ser difíciles de percibir para el ojo humano, por ejemplo, imágenes muy oscuras que no alcanzamos a ver bien, o imágenes muy claras. Adaptan la imagen de acuerdo a las necesidades requeridas de la intensidad.

### Vemos que si $\gamma =1$, entonces $s \propto r$.

### Si $\gamma =1=c$, $s=r$.

### También, si $\gamma >1$, la curva se caracteriza por ir sobre la recta $T(r)=r=s$, y si $\gamma <1$, se caracteriza por ir por debajo de esta recta. Esto es que.

### Lo anterior se traduce así: Cuando $\gamma =1=c$, los valores de intensidad se mantienen invariantes. Cuando $\gamma <1$ se aumentan las intensidades bajas de la imagen de entrada, por lo que se van aclarando las regiones más oscuras en la imagen de salida. La situación conversa ocurre para $\gamma >1$.

# Ejercicio:

### Implementar una transformación gamma a la imagen *"Luna.jpg"*. Vamos a fijar $c=1$ y variar $\gamma$ en dos valores, $\gamma _1 >1$ y $\gamma _2 <1$ 


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#Display de Luna.jpg
imagen=cv2.imread('luna.jpg',0)
plt.figure(figsize=(12,6))
plt.imshow(imagen, cmap = 'gray')
plt.title("Luna.jpg",fontsize=20)
plt.show()

#Transformación Gamma para gamma>1 y c=1, fijemos gamma=1.5:
c=1
gamma1=c*(imagen ** 1.5)  #La potencia arroja valores flotantes, queremos valores de formato uint8 (argumento abordado en la parte 2)
gamma1=np.array(gamma1,dtype=np.uint8)

plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.title("Imagen original", fontsize = 15)
plt.imshow(imagen, cmap = 'gray')
plt.subplot(1,2,2)
plt.title(r"Imagen de salida arrojada por la transformación gamma para $\gamma = 1.5$", fontsize = 15)
plt.imshow(gamma1, cmap = 'gray')
plt.show()

#Transformación Gamma para gamma<1 y c=1, fijemos gamma=0.3:
gamma2=c*(imagen ** 0.3) 
gamma2=np.array(gamma2, dtype = np.uint8)

plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.title("Imagen original", fontsize = 15)
plt.imshow(imagen, cmap = 'gray')
plt.subplot(1,2,2)
plt.title(r"Imagen de salida arrojada por la transformación gamma para $\gamma = 0.3$", fontsize = 15)
plt.imshow(gamma2, cmap = 'gray')
plt.show()
```
