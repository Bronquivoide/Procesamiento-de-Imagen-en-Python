### Extendiendo esta serie de notebooks, ahora nos corresponde abordar la transformación de autocontraste.

### Esta transformación se describe analíticamente mediante:

<font size="5"> $T(r)=255 \cdot \frac{r-r_{min}}{r_{max}-r_{min}}=s $ </font>

### En donde $0 \leq r \leq L-1$ es la intensidad asociada a los pixeles de la imagen de entrada y $ s $ la intensidad asociada a los pixeles de la imagen de salida.

### Esta transformación mapea el valor mínimo de una imagen de entrada al valor $0$ en la imagen de salida, y el nivel máximo de la imagen de entrada a $255$ en la de salida. Redistribuye linealmente los valores intermedios.

# Ejercicio:

### Implementar una transformación de autocontraste a la imagen*"Arroz.jpg"*.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#Display de Luna.jpg
imagen=cv2.imread('Arroz.jpg')
plt.figure(figsize=(12,6))
plt.imshow(imagen)
plt.title("Arroz.jpg",fontsize=20)
plt.show()

#Transformación de autocontraste con base al mínimo y al máximo:
print(f"La intensidad máxima asociada a la imagen de entrada es: {np.max(imagen)}")
print()
print(f"La intensidad mínima asociada a la imagen de entrada es: {np.min(imagen)}")

autocontraste=(255/(np.max(imagen) - np.min(imagen))) * (imagen - np.min(imagen))
autocontraste= np.array(autocontraste, dtype = np.uint8) #Float a uint8

plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.title("Imagen original", fontsize = 15)
plt.imshow(imagen, cmap = 'gray')
plt.subplot(1,2,2)
plt.title("Imagen de salida que arroja la transformación de autocontraste", fontsize = 15)
plt.imshow(autocontraste, cmap = 'gray')
plt.show()

print(f"La intensidad máxima asociada a la imagen de salida es: {np.max(autocontraste)}")
print()
print(f"La intensidad mínima asociada a la imagen de salida es: {np.min(autocontraste)}")
```
