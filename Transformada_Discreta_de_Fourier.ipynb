# Brevísima Introducción:

### Este notebook extiende la teoría de las transformaciones de Fourier sobre señales vista en el repositorio sobre procesamiento de señales (https://github.com/Bronquivoide/Signal-Proccesing-con-Python/tree/main), aplicándola ahora para el procesamiento de imágenes.

### En el procesamiento de señales abordamos las transformadas de Fourier para una señal representada analíticamente por una función unidimensional $f(t)$, ahora las abordaremos para una imagen representada por $f(x,y)$. Es decir, extenderemos su formulación y aplicación de sistemas unidimensionales a sistemas bidimensionales.

### Notemos que para las señales, el dominio original en el que son procesadas es el dominio de tiempo, por lo que una transformada de Fourier te lleva a describirla en su dominio recíproco, el dominio de frecuencias. Para el caso del procesamiento de imágenes, tenemos que el dominio original en el que se procesan, como hemos visto en todos los demás notebooks, es el dominio espacial, y las transformadas de Fourier las describirán en el dominio de frecuencias.

### La transformada de Fourier continua para una función arbitraria continua $f(x,y)$ está definida como:

<font size="5"> $ F(\omega , \phi)= \int _{-\infty}^{\infty} \int _{-\infty}^{\infty} f(x,y) \cdot exp[i2\pi (\omega x+ \phi y)] dydx $</font>

### O en términos de la identidad euleriana:

<font size="5"> $F(\omega , \phi)=\int _{-\infty}^{\infty} \int _{-\infty}^{\infty} f(x,y) \cdot [cos(2\pi(\omega x+ \phi y))-isen(2\pi(\omega x+ \phi y))] dydx$</font>

### Puede ser escrita como la suma de su parte real y la imaginaria, las cuales pueden ser vistas como funciones que reciben dos argumentos espaciales y mapean a una componente real y una imaginaria, denotadas de la siguiente manera:

<font size="5"> $ F(\omega , \phi)= Re(\omega , \phi)+iIm(\omega , \phi) \in \mathbb{C}^2 $</font>

### Cuyo módulo es:

<font size="5"> $ |F(\omega , \phi)|= \sqrt{Re ^2(\omega , \phi)+Im^2(\omega , \phi)} $</font>

### Y cuyo ángulo característico es:

<font size="5"> $ \theta (\omega , \phi) = tan ^{-1} (\frac{Im(\omega , \phi)}{Re(\omega , \phi)}) $</font>

### Por lo que la transformada de fourier inversa de caracter continuo (la que te lleva del dominio espacial al de frecuencias) es:

<font size="5"> $ F ^{-1}(\omega , \phi)=f(x,y)= \int _{-\infty}^{\infty} \int _{-\infty}^{\infty} f(x,y) \cdot exp[i2\pi (\omega x+ \phi y)] dydx $</font>

### En estos casos $\omega$ y $\phi$ son variables que representan frecuencias de espectro continuo.

### Continuando con la analogía del análisis de transformadas de Fourier en imágenes con el del análisis para señales, también en los algoritmos computacionales de procesamiento de imágenes puede ser útil implementar una transformada discreta, la cual está definida para una imagen arbitraria $f(x,y)$ de orden $M \times N$ como sigue:

<font size="5"> $ F(u , v)= \frac{1}{MN} \sum _{x=0}^{M-1} \sum _{y=0}^{N-1} f(x,y) \cdot exp[-i 2\pi (\frac{ux}{M}+\frac{vy}{N})], $</font>

<font size="5"> $\forall x \in [0,M-1] \subset \mathbb{Z}_0 ^+ , \; \forall y \in [0,N-1]\subset \mathbb{Z}_0 ^+$ </font>
 
### El kernel de esta transformada es la función $ g(x,y,u,v)=exp[-i 2\pi (\frac{ux}{M}+\frac{vy}{N})]$. Las variables $u,v$ son frecuencias de epectro discreto. 

### Y la transformada discreta de Fourier inversa es:

<font size="5"> $ f(x,y)= \sum _{x=0}^{M-1} \sum _{y=0}^{N-1} F(u , v) \cdot exp[i 2\pi (\frac{ux}{M}+\frac{vy}{N})]$  </font>

<font size="5"> $\forall u \in [0,M-1] \subset \mathbb{Z}_0 ^+ , \; \forall v \in [0,N-1]\subset \mathbb{Z}_0 ^+$ </font>

### En donde el kernel es la función $h(x,y,u,v)=exp[i 2\pi (\frac{ux}{M}+\frac{vy}{N})]$

### Las frecuencias normalizadas (ciclos por pixel) son $\nu _x = \frac{u}{M}$ y $\nu _y = \frac{v}{N}$

# Fundamentos del procesamiento de imágenes con la transformada discreta de Fourier: 

### Si consideramos a $f(x,y)$ como una imagen arbitraria de entrada, $h(x,y)$ un operador lineal y $g(x,y)$ una imagen de salida, recordemos que la convolución de $f$ con $h$ es:

<font size="5"> $g(x,y)= h(x,y) \diamond f(x,y)$</font>

### Con base a lo anterior, sea $G(u,v)$ la transformada de Fourier discreta de $g(x,y)$, $H(u,v)$ la de $h(x,y)$ y $F(u,v)$ la de $f(x,y)$. De acuerdo al teorema de la convolución tenemos que la transformada de Fourier discreta de la convolución en el dominio espacial entre $h(x,y)$ y $f(x,y)$, es el producto de las transformadas de Fourier discretas de cada una:

<font size="5"> $DFT[g(x,y)]= DFT[h(x,y)] \cdot DFT[f(x,y)]$</font>

<font size="5"> $\sim$</font>

<font size="5"> $G(u,v)= H(u,v)\cdot F(u,v)$</font>

### Por lo que en estos términos, la imagen procesada es:

<font size="5"> $g(x,y)=DFT ^{-1}[H(u,v)\cdot F(u,v)]$</font>

# Ejercicio:

### Implementar una transformada de Fourier a *Lena.tif* con OpenCV.



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#Display de la imagen original:
imagen_original=cv2.imread('Lena.tif', 0)
plt.figure(figsize=(12,6))
plt.imshow(imagen_original, cmap = 'gray')
plt.title(r"Imagen original, $f(x,y)$",fontsize=20)
plt.show()

#Cálculo de la DFT de la imagen:
DFT=cv2.dft(np.float32(imagen_original),flags=cv2.DFT_COMPLEX_OUTPUT) #cv2.DFT_COMPLEX_OUTPUT genera una DFT completa dando un complejo, descrito por la suma Re+iIm.
print(DFT.shape)
print()
M,N,componentes=DFT.shape
print(f"El orden de la transformada de Fourier discreta es de M={M} x N={N}")
print()
print(f"La DFT está descrita en {componentes} componentes: una componente real y una imaginaria.")
print()

#Reordenando la DFT moviendo las componentes de frecuencia cero al centro del arreglo asociado a la imagen para una mejor visualización del espectro:
DFT_ordenada=np.fft.fftshift(DFT) #Sin esta organización se trabaja desde la esquina superior izquierda del arreglo asociado a la imagen.

#Tomando el módulo del espectro:
módulo=20*np.log(cv2.magnitude(DFT_ordenada[:,:,0],DFT_ordenada[:,:,1]) ) #Se toma la magnitud aislando en un comando las dos componentes, la real y la imaginaria.
#Este es un procedimiento comunmente utilizado al aplicar una transformada de Fourier en una imagen. 
#El término logarítmico permite visualizar mejor las frecuencias bajas y las altas.
M1,N1=módulo.shape
print(f"La estructura del arreglo correspondiente al espectro con base al módulo es de M={M1} x N={N1} ")

#Comparativa imagen de entrada y su correspondiente magnitud del espectro de frecuencias generada por la DFT:
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.imshow(imagen_original, cmap = 'gray')
plt.title('Imagen de entrada',fontsize=20)
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(módulo, cmap = 'gray')
plt.title('Espectro de Fourier discreto',fontsize=20)
plt.xticks([])
plt.yticks([])
plt.show()
```
